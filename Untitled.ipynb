```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import datetime as dt
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("Electricity Consumption.csv")
data
```

```python
data['DATE']=pd.to_datetime(data['DATE'])
```

```python
data.isna().sum()
```

```python
data['Electricty_Consumption_in_TW'].plot()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
datapoint=adfuller(data['Electricty_Consumption_in_TW'],maxlag=15)
labels=['test_parameter',"p-value","lagused","dataobs"]
d=dict(zip(labels,datapoint[:4]))
d
```

```python
if d["p-value"]>0.05:
    print("non")
else:
    print("sationary")
```

```python
data["lag"]=data["Electricty_Consumption_in_TW"]-data["Electricty_Consumption_in_TW"].shift(1)
datapoint=adfuller(data["lag"].dropna())
d=dict(zip(labels,datapoint))
d
```

```python
if d['p-value']>0.05:
    print("not")
else:
    print("stationary")
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sn
```

```python
data1=data[["Electricty_Consumption_in_TW"]]
data1.set_index(data["DATE"],inplace=True)
```

```python
fig=sn.graphics.tsa.plot_acf(data1['Electricty_Consumption_in_TW'],lags=70)
```

```python
fig=sn.graphics.tsa.plot_pacf(data1['Electricty_Consumption_in_TW'],lags=50)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
results=seasonal_decompose(data1["Electricty_Consumption_in_TW"])
plt.figure(figsize=(12,8))
plt.subplot(4,1,1)
plt.plot(results.observed)
plt.title("observed")
plt.subplot(4,1,2)
plt.plot(results.trend)
plt.title("trend")
plt.subplot(4,1,3)
plt.plot(results.seasonal)
plt.title("seasonal")
plt.subplot(4,1,4)
plt.plot(results.resid)
plt.title("residual")
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model=ARIMA(data1['Electricty_Consumption_in_TW'],order=(15,0,5))
model_pred=model.fit()
print(model_pred.summary())
```

```python
pred=pd.Series(model_pred.predict(start="1973-01-01",end="2019-09-01"))
pred
```

```python
arima=pd.concat([data1,pred],axis=1)
arima[["Electricty_Consumption_in_TW","predicted_mean"]].plot()
arima
```

```python
forecasted = model_pred.get_forecast(steps=24)
forecasted = forecasted.predicted_mean
forecasted
```

```python
plt.plot(arima['Electricty_Consumption_in_TW'])
plt.plot(arima['predicted_mean'])
plt.plot(forecasted)
plt.xlabel('Date')
plt.ylabel('Electricity Consumption')
plt.title('Arima Model Electricity Consumption Forecast')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_percentage_error(arima.Electricty_Consumption_in_TW,arima.predicted_mean)
metrics.mean_squared_error(arima.Electricty_Consumption_in_TW,arima.predicted_mean,squared=False)
np.sqrt(np.mean(np.square((arima["Electricty_Consumption_in_TW"]-arima["predicted_mean"])/arima["Electricty_Consumption_in_TW"])))*100
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model=SARIMAX(data1["Electricty_Consumption_in_TW"],order=(15,0,5),seasonal_order=(0, 0, 1, 12))
model_pred=model.fit()
print(model_pred.summary())
```

```python
pred=pd.Series(model_pred.predict(start="1973-01-01",end="2019-09-01"))
pred
```

```python
sarima=pd.concat([data1,pred],axis=1)
sarima[["Electricty_Consumption_in_TW","predicted_mean"]].plot()
sarima
```

```python
forcast=model_pred.forecast(steps=24)
forcast
```

```python
plt.plot(sarima["Electricty_Consumption_in_TW"])
plt.plot(sarima["predicted_mean"])
plt.plot(forcast)
```

```python
metrics.mean_absolute_percentage_error(sarima["Electricty_Consumption_in_TW"],sarima["predicted_mean"])
```

```python
metrics.mean_squared_error(sarima["Electricty_Consumption_in_TW"],sarima["predicted_mean"],squared=False)
```

```python
np.sqrt(np.mean(np.square((sarima["Electricty_Consumption_in_TW"]-sarima["predicted_mean"])/sarima["Electricty_Consumption_in_TW"])))*100
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
model=ExponentialSmoothing(data1["Electricty_Consumption_in_TW"],trend="add",seasonal="add",seasonal_periods=12)
model_pred=model.fit()
print(model_pred.summary())
```

```python
ets=model_pred.predict(start="1973-01-01",end="2019-09-01")
ets=pd.concat([data1,ets],axis=1)
ets[["Electricty_Consumption_in_TW",0]].plot()
ets
```

```python
forcast=model_pred.forecast(steps=24)
forcast
```

```python
plt.plot(ets[["Electricty_Consumption_in_TW",0]])
plt.plot(forcast)
```

```python
metrics.mean_absolute_percentage_error(ets["Electricty_Consumption_in_TW"],ets[0])
```

```python
metrics.mean_squared_error(ets["Electricty_Consumption_in_TW"],ets[0],squared=False)
```

```python
np.sqrt(np.mean(np.square((ets["Electricty_Consumption_in_TW"]-ets[0])/ets["Electricty_Consumption_in_TW"])))*100
```
